## 1) Introdução

Assim como no caso discreto, podemos calcular probabilidades, quantis e simular dados de uma v.a. contínua. Seja ```nome``` o rótulo que o R utiliza para identificar uma distribuição e ```parametros``` os parâmetros dessa distribuição. Por exemplo, para uma v.a. binomial, ```nome``` era igual a ```binom``` e ```parametros``` eram os valores ```n,p```. Utiliza-se a mesma sintaxe básica:

* ```dnome(x, parametros)```: retorna valores da f.d.p. de uma v.a. $X$ no ponto $x$;
* ```pnome(x, parametros)```: retorna valores da f.d.a. de $X$ no ponto $x$;
* ```qnome(alpha, parametros)```: retorna o quantil de ordem $0\leq\alpha\leq 1$ da distribuição de $X$. Ou seja, o valor de $x$ tal que $F(x)=\alpha$;
* ```rnome(N,parametros)```: gera uma amostra de tamanho ```N``` da v.a. $X$.

## 2) Distribuição Qui-quadrado

A distribuição **qui-quadrado** (chi-square em inglês) é uma distribuição muito importante que surge na etapa inferencial de algumas análises estatísticas como veremos posteriormente.

Sejam $X_1,X_2,\ldots,X_k$ v.a.'s independentes, todas com distribuição $N(0,1)$. Uma v.a. $Q$ com distribuição qui-quadrado surge da seguinte relação:

$$Q=X_1^2+X_2^2+\cdots+X_k^2.$$

Neste caso, dizemos que $Q$ possui distribuição qui-quadrado com $k$ **graus de liberdade**. **Notação:** $Q\sim\chi_k^2.$

Sua f.d.p. é dada por

$$f(x)=\frac{1}{2^{k/2}\Gamma(k/2)}x^{\frac{k}{2}-1}e^{-\frac{x}{2}}, \ \ \ k>0, x>0.$$

O símbolo $\Gamma(\cdot)$ denota a **função gama**. A figura a seguir ilustra o gráfico dessa f.d.p. variando o valor de $k$.



Além disso, se $Q\sim\chi_k^2$, mostra-se que $\mathbb{E}(Q)=k$ e $\mbox{Var}(Q)=2k$.

Para uma distribuição qui-quadrado com $k$ graus de liberdade, temos as seguintes funções implementadas no R:

* ```dchisq(x, k)```: retorna valores da f.d.p. no ponto $x$;
* ```pchisq(x, k)```: retorna valores da f.d.a. no ponto $x$;
* ```qchisq(alpha, k)```: retorna o quantil de ordem $0\leq\alpha\leq 1$ da distribuição. Ou seja, o valor de $x$ tal que $F(x)=\alpha$;
* ```rchisq(N, k)```: gera uma amostra de tamanho ```N``` com distribuição qui-quadrado com $k$ graus de liberdade.

## 3) Distribuição t-Student

Sejam $Z\sim N(0,1)$ e $Q\sim\chi_k^2$, mostra-se que a v.a.

$$T=\frac{Z}{\sqrt{\frac{Q}{k}}},$$

é também uma variável aleatória cuja distribuição recebe o nome de t-Sudent com **$k$ graus de liberdade**. **Notação:** $T\sim t_k$.

A distribuição **t-Student**, assim como a distribuição anterior, surge frequentemente em problemas de inferência estatística. Ela possui caudas mais pesadas que as da distribuição normal, o que faz com que essa distribuição seja mais adequada para lidar com valores discrepantes.

Sua f.d.p. é dada por

$$f(x)=\frac{\Gamma(\frac{k+1}{2})}{\sqrt{k\pi}\Gamma(k/2)}\left(1+\frac{x^2}{k}\right)^{-\frac{k+1}{2}}, \ \ \ k>0,x\in\mathbb{R}.$$

Novamente, $\Gamma(\cdot)$ denota a função gama. A figura a seguir ilustra o gráfico dessa f.d.p. variando o valor de $k$.



A esperança e a variância de uma v.a. t-Student só está definida para certos valores de $k$. Logo, se $T\sim t_k$, temos que

$$\mathbb{E}(T)=0, \ \ \ \mbox{se} \ \ \ k>1,$$

$$\mbox{Var}(T)=\frac{k}{k-1}, \ \ \ \mbox{se} \ \ \ k>2.$$

Se $k$ for suficientemente grande ($k\rightarrow\infty$), temos que $T$ se aproxima da distribuição $N(0,1)$.

Para uma distribuição t-Student com $k$ graus de liberdade, temos as seguintes funções implementadas no R:

* ```dt(x, k)```: retorna valores da f.d.p. no ponto $x$;
* ```pt(x, k)```: retorna valores da f.d.a. no ponto $x$;
* ```qt(alpha, k)```: retorna o quantil de ordem $0\leq\alpha\leq 1$ da distribuição. Ou seja, o valor de $x$ tal que $F(x)=\alpha$;
* ```rt(N, k)```: gera uma amostra de tamanho ```N``` com distribuição qui-quadrado com $k$ graus de liberdade.
